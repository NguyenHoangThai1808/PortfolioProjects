Xây dựng mô hình tính lợi nhuận (y) dựa trên chi phí quản lý (x)

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(6,4),'figure.dpi':120})
```

```python
link='https://drive.google.com/file/d/1kP7c_onaCSo6uDfzZDW0QAz9oK-9b58C/view?usp=sharing'
```

```python
path = 'https://drive.google.com/uc?export=download&id='+link.split('/')[-2]
df = pd.read_csv(path)
```

```python
df.head(10)
```

```python
plt.xlabel('Chi phí RnD (1000$)',fontsize=12)
plt.ylabel('Chi phí quản lí (1000$)',fontsize=12)
plt.grid()
plt.scatter(df.RnD, df.Adm, color='brown', marker='o')
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
model_lin=linear_model.LinearRegression()
model_lin.fit(df[['RnD']],df.Adm)
```

```python
print('a=',model_lin.coef_)
print('b=',model_lin.intercept_)
```

```python
y_pred=model_lin.predict(df[['RnD']])
y_pred
```

```python
plt.xlabel('Chi phí RnD (1000$)',fontsize=12)
plt.ylabel('Chi phí quản lí (1000$)',fontsize=12)
plt.scatter(df.RnD, df.Adm, color='green', marker='o')
plt.plot(df.RnD,y_pred,color='red')
plt.legend(['y dự báo'])
plt.grid()
```

```python
from sklearn.metrics import r2_score
ac = r2_score(df.Adm, y_pred)
print('The accuracy of this model is: R-square=',"{:.2f}".format(ac))
```
